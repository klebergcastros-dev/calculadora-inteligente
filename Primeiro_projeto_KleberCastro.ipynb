```python
"""
Projeto: Calculadora Inteligente
Autor: Kleber Castro
Descrição:
Este programa permite que o usuário realize operações matemáticas básicas
(soma, subtração, multiplicação e divisão).

Requisitos atendidos:
- Uso de input para receber dados
- Conversão para float
- 4 operações matemáticas
- Estrutura condicional (if/elif/else)
- Laço de repetição (while)
- Exibição do resultado com print
- Conversão do resultado final para número inteiro (int)
"""

print("=== CALCULADORA INTELIGENTE ===")

# Loop principal do programa
# O programa continuará executando até o usuário decidir sair
while True:

    # ==================================================
    # 1. Entrada de Dados
    # ==================================================

    # Solicita dois números ao usuário
    # A conversão para float permite aceitar números decimais
    primeiro_numero = float(input("Digite o primeiro número: "))
    segundo_numero = float(input("Digite o segundo número: "))

    # ==================================================
    # 2. Menu de Operações
    # ==================================================

    print("\nEscolha a operação desejada:")
    print("1 - Soma")
    print("2 - Subtração")
    print("3 - Multiplicação")
    print("4 - Divisão")

    # Recebe a opção escolhida pelo usuário
    opcao_operacao = input("Digite o número da operação: ")

    # ==================================================
    # 3. Processamento das Operações
    # ==================================================

    if opcao_operacao == "1":
        # Soma dos números
        resultado_operacao = int(primeiro_numero + segundo_numero)
        print("Resultado da soma:", resultado_operacao)

    elif opcao_operacao == "2":
        # Subtração dos números
        resultado_operacao = int(primeiro_numero - segundo_numero)
        print("Resultado da subtração:", resultado_operacao)

    elif opcao_operacao == "3":
        # Multiplicação dos números
        resultado_operacao = int(primeiro_numero * segundo_numero)
        print("Resultado da multiplicação:", resultado_operacao)

    elif opcao_operacao == "4":
        # Antes de dividir, verificamos se o divisor é diferente de zero
        if segundo_numero != 0:
            resultado_operacao = int(primeiro_numero / segundo_numero)
            print("Resultado da divisão:", resultado_operacao)
        else:
            # Tratamento de erro para divisão por zero
            print("Erro: Não é possível dividir por zero.")

    else:
        # Caso o usuário digite uma opção inválida
        print("Opção inválida. Tente novamente.")

    # ==================================================
    # 4. Controle de Continuidade
    # ==================================================

    # Pergunta ao usuário se deseja continuar usando a calculadora
    continuar_programa = input("\nDeseja realizar outra operação? (s/n): ")

    # Se a resposta for diferente de "s", o programa será encerrado
    if continuar_programa.lower() != "s":
        print("Encerrando a calculadora. Até mais!")
        break
```
